# Three Methods to cover

* df.reset_index()
* df.set_index()
* df.sort_index()

# df.reset_index()

```python
df_arg = df[df['job_country'] == 'Argentina']
```

```python
df_arg.head(5)
```

```python
df_arg.reset_index(inplace=True)

df_arg
```

# df.set_index()

```python
df_arg.set_index('job_index', inplace=True)

df_arg
```

# df.sort_index()

```python
median_pivot = df_arg.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc=['median', 'min', 'max'])

median_pivot
```

```python
median_pivot.sort_values(by=('median', 'salary_year_avg'), ascending=False)
```

```python
median_pivot.sort_index()
```

```python
median_pivot.sort_index(inplace=True)

median_pivot
```
