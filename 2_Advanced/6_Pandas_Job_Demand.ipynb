# Exercise

* Use pivots to manipulates indexes so you can agregate months in order.
* Filter the data and select the top 3 most usual jobs in Argentina.


```python
# import libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# cleanng data
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

We will copy the dataframe in a new variable

```python
df_arg = df[df['job_country'] == 'Argentina'].copy()
```

Check it

```python
df_arg
```

Let's create a new column calles 'job_posted_month' from the already existing 'job_posted_date' using the '.strftime()' method.

Later we check.

```python
df_arg['job_posted_month'] = df_arg['job_posted_date'].dt.strftime('%B')

df_arg
```

We use the '.pivot_table()' method to reorganize the dataframe

```python
df_arg.pivot_table(
    index='job_posted_month',
    columns='job_title_short',
    aggfunc='size',
)
```

And save it to a new varialbe

```python
df_arg_pivot = df_arg.pivot_table(
                                    index='job_posted_month',
                                    columns='job_title_short',
                                    aggfunc='size',
                                )
```

The problem now is that this dataframe is not in historical order. Rather it's in alphabetical order, we don't want this

We create a new column with the '.reset_index()' method

```python
df_arg_pivot.reset_index()
```

We use the Pandas '.to_datetime()' method and the '.dt.month' method to asign an index to every month

```python
df_arg_pivot.reset_index(inplace=True)

df_arg_pivot['month_number'] = pd.to_datetime(df_arg_pivot['job_posted_month'], format='%B').dt.month

df_arg_pivot['month_number']
```

```python
df_arg_pivot
```

Now we use the '.sort_values()' method to sort them in historical order

```python
df_arg_pivot.sort_values('month_number')
```

We repeat, modifing the dataframe with 'inplace=', and then we drop the numerical column created before

```python
df_arg_pivot.sort_values('month_number', inplace=True)

df_arg_pivot.set_index('job_posted_month', inplace=True)
```

```python
df_arg_pivot.drop(columns='month_number', inplace=True)
```

```python
df_arg_pivot
```

Now we can plot it directly with 'plot()'

```python
df_arg_pivot.plot(kind='line')
```

We create a new variable named 'top_3', and using 'value_counts()' and 'head(3)' we decide to drop everything but the top 3 most usual jobs in Argentina

```python
top_3 = df_arg['job_title_short']. value_counts().head(3)
top_3 = top_3.index.tolist()
top_3
```

We plot it as before

```python
df_arg_pivot[top_3].plot(kind='line')
```

We complete the plot using the matplotlib.pyplot library

```python
df_arg_pivot[top_3].plot(kind='line')
plt.title('Monthly Job Postings for Top Data Jobs in Argentina')
plt.xlabel('2023')
plt.ylabel('Job count')
plt.legend()
plt.show()
```
