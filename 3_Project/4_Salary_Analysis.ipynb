# How well do jobs and skills pay for Data Analysts?

Methodology
1. Evaluate median salary for top 6 data jobs
2. Find median salary per skill for Data Analysts
3. Visualize for highest paying skills and most demanded skills

```python
# import libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import seaborn as sns

# loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# cleanng data
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

# cleaning the 'job_skills' into a list datatype
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
# filter for the job and the country

df_ARG = df[(df['job_country'] == 'Argentina')].dropna(subset=['salary_year_avg'])
```

```python
df_ARG
```

```python
# let's get the top 6 jobs
job_titles = df_ARG['job_title_short'].value_counts().index[:6].tolist()

job_titles
```

```python
# filtering to see if those titles are in the top 6
df_ARG_top6 = df_ARG[df_ARG['job_title_short'].isin(job_titles)]

df_ARG_top6
```

```python
# we need to reorder by the median salary, so they appear in the right order in the plot
df_ARG_top6.groupby('job_title_short')['salary_year_avg'].median().sort_values(ascending=False).index

# groupby to find out what are the diferent median salaries and them order by that
# median cause we want the median salary
# sort_values, ascending=False cause we want it in descending order
# we dont care what the median values are, we just want the index
```

```python
# we save it in a new variable
job_order = df_ARG_top6.groupby('job_title_short')['salary_year_avg'].median().sort_values(ascending=False).index
```

```python
# now let's plot with boxplot
# we add the 'order' parameter, using the last defined variable
sns.boxplot(data=df_ARG_top6, x='salary_year_avg', y='job_title_short', order=job_order)
sns.set_theme(style='ticks')

plt.title('Salary distribution in Argentina')
plt.xlabel('Yearly Salary (U$D)')
plt.ylabel('')
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${int(x/1000)}K'))
plt.xlim(0, 250000)
plt.show()
```

# Investigate Median Salary vs Skill for Data Analysts

```python
# let's create a variable that saves tha data analysts jobs in Argentina
df_DA_ARG = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'Argentina')].copy()

# and let's drop any  NaN values
df_DA_ARG = df_DA_ARG.dropna(subset=['salary_year_avg'])
```

```python
# now we explode the 'job_skills' column
df_DA_ARG = df_DA_ARG.explode('job_skills')

# and let's look at the 'salary_year_avg' and 'job_skills' columns
df_DA_ARG[['salary_year_avg', 'job_skills']].head(5)
```

```python
# let's find the top paying skills with groupby
# and then we'll sort them
df_DA_ARG.groupby('job_skills')['salary_year_avg'].agg(['count', 'median']).sort_values(by='median', ascending=False)
```

```python
# now we save it to a new variable
df_DA_ARG_pay = df_DA_ARG.groupby('job_skills')['salary_year_avg'].agg(['count', 'median']).sort_values(by='median', ascending=False)
```

```python
df_DA_ARG_pay
```

```python
# now we create a variable similar to last, but sorting by count
df_DA_ARG_skills = df_DA_ARG.groupby('job_skills')['salary_year_avg'].agg(['count', 'median']).sort_values(by='count', ascending=False)
```

```python
df_DA_ARG_skills
```

```python
fig, ax = plt.subplots(2, 1)

sns.set_theme(style='ticks')

# Top 10 Highest Paid Skills for Data Analysts
sns.barplot(data=df_DA_ARG_pay, x='median', y=df_DA_ARG_pay.index, ax=ax[0], hue='median', palette='dark:b_r')
ax[0].legend().remove()

# df_DA_ARG_pay[::-1].plot(kind='barh', y='median', ax=ax[0], legend=False)
ax[0].set_title('Top 10 Highest Paid Skills for Data Analysts')
ax[0].set_xlabel('')
ax[0].set_ylabel('')
ax[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))

# Top 10 Most In-Demand Skills for Data Analysts
sns.barplot(data=df_DA_ARG_skills, x='median', y=df_DA_ARG_pay.index, ax=ax[1], hue='median', palette='light:b')
ax[1].legend().remove()

# df_DA_ARG_skills[::-1].plot(kind='barh', y='median', ax=ax[1], legend=False)
ax[1].set_title('Top 10 Most In-Demand Skills for Data Analysts')
ax[1].set_xlabel('Median Salary (USD)')
ax[1].set_ylabel('')
ax[1].set_xlim(ax[0].get_xlim())  # Set the same x-axis limits as the first plot
ax[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))

plt.tight_layout()
plt.show()
```
